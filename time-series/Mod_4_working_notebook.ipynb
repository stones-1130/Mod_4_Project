```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pandas.plotting import autocorrelation_plot, lag_plot
import matplotlib as mpl
```

```python
#CODE ADAPTED FROM JAMES' LECTURE FROM FULL TIME COHORT YOUTUBE VIDEO

def plot_acf_pacf(df, zipcode, figsize=(10,8), lags=24): 
    
    '''Plot the ACF and PACF FOR ONE ZIPCODE USING A DIFFERENCING OF 1'''
    
    #BUILD SUBPLOTS
    fig,ax = plt.subplots(nrows=3, figsize=figsize)
    
    #SELECT ONE ZIPCODE FROM THE DATAFRAME OF ALL ADA COUNTY ZIPCODES
    zipcode = zipcode
    
    #COPY THE DATAFRAME AND USE A DIFFERENCING OF 1 WHILE DROPPING ANY NA VALUES
    ts = df[zipcode].copy().diff(1).dropna(0)
    ts.plot(ax=ax[0])
    
    
    #PLOT ACF AND PACF
    plot_acf(ts, ax=ax[1], lags=lags, fft=True)
    plot_pacf(ts, ax=ax[2], lags=lags)
    fig.tight_layout()
    
    fig.suptitle(f"Zipcode: {ts.name}" , y=1.1, fontsize=20)
    
    #SET NUMBER OF LAGS AS THE X-AXIS FOR THE ACF AND PACF PLOTS
    for a in ax[1:]:
        a.xaxis.set_major_locator(mpl.ticker.MaxNLocator(min_n_ticks=lags, integer=True))
        a.xaxis.grid()
    return fig,ax
    

plot_acf_pacf(df2, 83616, lags=15);
```

```python
# TRAIN TEST SPLIT INDEX
train_size = 0.8
split_idx = round(len(ts) * train_size)
split_idx

# SPLIT
train = ts.iloc[:split_idx]
test = ts.iloc[split_idx:]
```

```python
#VISUALIZE TRAIN TEST SPLIT
fig, ax = plt.subplots()
kws = dict(ax=ax,marker='o')
train.plot(**kws)
test.plot(**kws)
ax.legend(bbox_to_anchor=[1,1]);
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn import metrics
```

```python
#PLOT PREDICTION RESULTS

def plot_train_test_pred(train, test, pred_df):
    fig, ax = plt.subplots()
    kws = dict(marker='o')
    
    ax.plot(train, label='Train', **kws)
    ax.plot(test, label='Test', **kws)
    ax.plot(pred_df['prediction'], label='Prediction', ls='--', **kws)
    ax.set(ylabel="Median price - $ USD", xlabel='Date')
    
    ax.fill_between(x=pred_df.index, y1=pred_df['lower'], y2=pred_df['upper'])
    ax.legend(bbox_to_anchor=[1,1])
    fig.tight_layout()
    return fig,ax
```

```python
#AUTO-ARIMA IS BASICALLY A GRIDSEARCH FOR TIME SERIES
from pmdarima.arima import auto_arima

auto_model = auto_arima(train, start_p=0, start_q=0)
display(auto_model.summary())

auto_model.plot_diagnostics();
```

```python
## Baseline SARIMAX model
p = d = q = 1
model = SARIMAX(train,order=(p,d,q),).fit()
```

```python
#BUILD LOOP FOR FORCASTING PRICES

forecast = model.get_forecast(len(test))

def forecast_to_df(forecast, zipcode):
    
    test_pred = forecast.conf_int()
    test_pred[zipcode] = forecast.predicted_mean
    test_pred.columns = ['lower', 'upper', 'prediction']
    return test_pred

pred_df = forecast_to_df(forecast, zipcode)
pred_df
```

```python
#USE DYNAMIC FORCASTING TO PREDICT HOUSE PRICES 5 YEARS INTO THE FUTURE

#CREATE AN EMPTY DICTIONARY TO STORE THE MODELING RESULTS
RESULTS = {}

for zipcode in zipcode_list:
    '''Forecast mean home values for each Boise/Ada County zipcode
    from May 2018 - April 2023 '''
    
    print("ZIPCODE: ", zipcode)

    # MAKE EMPTY DICTIONARY FOR ZIPCODE DATA
    zipcode_z = {}

    # COPY TIME SERIES DATAFRAME
    ts = df2[zipcode].copy()

    # TRAIN TEST SPLIT INDEX
    train_size = 0.85
    split_idx = round(len(ts) * train_size)
    split_idx

    # SPLIT
    train = ts.iloc[:split_idx]
    test = ts.iloc[split_idx:]

    # GET BEST PARAMS USING AUTO ARIMA
    gs_model = auto_arima(train, start_p=0, start_q=0)
    
    #MODEL USING SARIMAX
    best_model = SARIMAX(train, order=gs_model.order, 
                         seasonal_order=gs_model.seasonal_order).fit()

    # GET PREDICTIONS FOR 5 YEARS IN ADVANCE (2018-2023)
    ## DATA IS SAMPLED MONTHY. THEREFORE 5-YEARS IN THE FUTURE IS 12x5 = 60 MONTHS(steps)
    pred = best_model.get_forecast(steps=60)
    pred_df = forecast_to_df(pred, zipcode)

  
    # SAVE INFO TO DICTIONARY
    zipcode_z['pred_df'] = pred_df
    zipcode_z['model'] = best_model
    zipcode_z['train'] = train
    zipcode_z['test'] = test

    # DISPLAY RESULTS
    display(best_model.summary())
    plot_train_test_pred(train, test, pred_df)
    plt.show()

    # SAVE ZIPCODE DICTIONARY IN RESULTS DICTIONARY FOR FUTURE USE

    RESULTS[zipcode] = zipcode_z
    print('---'*30, end='\n\n')

```
